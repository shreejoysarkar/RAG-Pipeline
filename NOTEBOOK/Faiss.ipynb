Facebook AI similariy search (FAISS) is a library for efficient similarity search and clustering of dense vectors. It implements a wide range of algorithms for similarity search, including brute-force search, LSH (Locality Sensitive Hashing), and various tree-based methods. FAISS is designed to be fast and scalable, making it suitable for large-scale applications in fields such as information retrieval, computer vision, and natural language processing.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS

from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("../DATA/text_files/RAG.txt")
documents = loader.load()

text_splitter = CharacterTextSplitter(
    chunk_size = 200, chunk_overlap = 30
)

docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="gemma:2b")
db = FAISS.from_documents(documents, embeddings)

```

```python
db
```

```python
## querying

query = "who authored Retrieval-Augmented Generation for Knowledge-Intensive NLP Task? "

docs = db.similarity_search(query)
docs
```

```python
docs[0].page_content
```

## As a Retriever


We can also convert the vectorstore into a Retriever class. This allows us to easily use it in other LangChain methods, which largely work with retrievers.


```python
retriever = db.as_retriever()
retriever.invoke(query)
```

## Similarity Search with score

There are some FAISS specific methods. One of them is similarity_search_with_score, which allows you to return not only the documents but also the distance score of the query to them. The returned distance score is L2 distance. Therefore, a lower score is better.

```python
docs_and_scores = db.similarity_search_with_score(query)
docs_and_scores
```
